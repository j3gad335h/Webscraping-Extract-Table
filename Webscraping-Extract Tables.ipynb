```python
#Import libraries
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

#  Check whether web page can be scraped or not

```python
url='https://en.wikipedia.org/wiki/Economy_of_India'
page=requests.get(url)
if str(page) == '<Response [200]>':
    print("Webpage can be scraped")
else:
    print("Sorry webpage can't be scraped")
```

#  Extract all tables using bs4

```python
soup=BeautifulSoup(page.text,'html.parser') #convert page to soup format
table=soup.find_all('table') #find all tags that contains 'table'
tab = soup.find_all("table",{"class":"wikitable"}) #filter tages that contains class wikitable
```

```python
tab[0]
```

# Extract headers from table

```python
header=[]
for i in tab[0].find_all('th'): #th-table header
    title=i.text
    header.append(title)
print(header)
```

```python
header[0][-1:]
```

```python
#Lets remove \n from headername
new_header=[]
for i in range(0,len(header)):
    if header[i][-1:]=='\n':
        h=header[i][0:-1]
    else:
        h=header[i]
    new_header.append(h)
print(new_header)
```

#  Create dataframe using headers as column name

```python
my_data=pd.DataFrame(columns=new_header)
#create a for loop and extract all data fom each row
for j in tab[0].find_all('tr')[1:]:
    row_data=j.find_all('td')
    row=[i.text for i in row_data]
    length=len(my_data)
    my_data.loc[length]=row
    
```

```python
my_data
```

```python
#lets remove all \n from datas
my_data = my_data.replace(r'\n','', regex=True) 
```

```python
my_data
```

```python
file_name='India GDP.xlsx'
my_data.to_excel(file_name)
print("DataFrame converted to excel successfully")
```

#  Convert the code into function

```python
def get_url(url):
    page=requests.get(url)
    if str(page) == '<Response [200]>':
        print(f"'{url}' can be scraped")
    else:
        print(f"Sorry '{url}' can't be scraped")
```

```python
url='https://www.worldometers.info/coronavirus/'
get_url(url)
```

```python
def list_tables(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.text,'html.parser') #convert page to soup format
    table=soup.find_all('table') #find all tags that contains 'table'
    table_class=[]
    for j in range(0,len(table)):
        s=""
        x=str(table[j])
        for i in range(0,len(x)):
            if x[i] != '>':
                s=s+x[i]
            elif  x[i] == '>':
                break
        table_class.append(s)
    return table,table_class
    
```

```python
table,table_class=list_tables(url)
table_class
```

# Create dataframe function

```python
def create_data(table):
    headers = []
    for i in table.find_all('th'):
        title = i.text
        headers.append(title)
    mydata = pd.DataFrame(columns = headers)
    # Create a for loop to fill mydata
    for j in table.find_all('tr')[1:]: #[0:]-->Header row
        row_data = j.find_all('td')
        row = [i.text for i in row_data]
        length = len(mydata)
        mydata.loc[length] = row
    return mydata
```
